```python
# this is for my own recollection for visual exploring data with python. Signal plotting matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import mpld3
mpld3.enable_notebook()
```

```python
# generating test signal, sum of various sinusoidal at different sequences with exp decay func
fs = 60
ts = np.arange(0,60,1/fs)
len(ts)
frequencies = [2,0.2, 1, 2,5,8]
amps = [5,2,1,0.1,5]
sig  = np.sum([a*np.sin(ts*2*np.pi*f) for a, f in zip(amps, frequencies)], axis=0)*np.exp(-0.05 * ts)
```

```python
plt.figure(figsize=(12,3))
plt.plot(ts,sig)
```

```python
# zero crossing y axis
zero_crossings = np.where(np.diff(np.sign(sig)))[0]
zero_crossings
```

```python
#The zero-crossings we found aren't exactly at 0. 
#This is because our original signal isn't continuous; 
#it's digitally sampled! That means we didn't sample the 
#signal exactly when it crossed zero and our zero-crossings 
#are the closest point we have to a "real" zero-crossing.

#We can visualize this by only plotting the signal points 
#that we have and not connecting them with lines in between.
plt.figure(figsize=(12,3))
plt.plot(ts,sig, 'b.')
plt.plot(ts[zero_crossings],sig[zero_crossings], 'r.', ms=10)
```
